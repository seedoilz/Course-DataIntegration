```python
import pandas as pd
```

```python
df = pd.read_csv("./star_original.csv")
df
```

```python
for column in df.columns.to_list():
    print(column,":")
    missing_count = df[column].isnull().sum()
    # print(missing_count)
    total_count = 290658
    # print(total_count)
    missing_percent = missing_count / total_count
    print(missing_percent)
    if(missing_percent>0.7):
        df = df.drop(column,axis=1)
        print("deleted\n")
```

```python
for column in df.columns.to_list():
    zero_count = len(df[df[column] == 0])
    print('%s, 零值：%d' % (column, zero_count))
    missing_percent = zero_count / 290658
    if(missing_percent > 0.7):
        df.drop(column,axis=1,inplace=True)
        print("deleted\n")
```

```python
for column in df.columns.to_list():
    print(column)
    zero_count = len(df[df[column] == 0])
    null_count = df[column].isnull().sum()
    missing_count = zero_count + null_count
    missing_percent = missing_count / 290658
    if(missing_percent > 0.7):
        df.drop(column,axis=1,inplace=True)
        print("deleted\n")
```

```python
for column in df.columns.to_list():
    if df[column].dtype == 'object':
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        df[column].fillna(df[column].median(), inplace=True)
df
```

```python
import pandas as pd
```

```python
df = pd.read_csv("./star_fillna.csv")
for column in df.columns.to_list():
    if column == 'star_level':
        print("true")
```

```python
for column in df.columns.to_list():
#     print(column)
    if df[column].dtype != 'object' and column != 'star_level':
        print(column)
        if(column == 'star_level'):
            print("error")
        if(column == 'uid'):
            print("error")
        low = df[column].quantile(0.05)
        high = df[column].quantile(0.95)
        # 使用clip方法替换小于5%和大于95%的值
        df[column] = df[column].clip(lower=low, upper=high)
```

```python
df.drop(columns='Unnamed: 0.1',axis=1,inplace=True)
df.drop(columns='Unnamed: 0',axis=1,inplace=True)
df
```

```python
df.to_csv("./star_error_data.csv")
```
