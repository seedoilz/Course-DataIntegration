```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("./star_train.csv")
df = df.drop("Unnamed: 0",axis=1)
df = df.drop("Unnamed: 0.1",axis=1)
# 定义特征和目标
X = df.drop("star_level", axis=1)
X = X.drop("uid", axis=1)
y = df["star_level"]

# 分割数据集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=14)
```

```python
rf = RandomForestClassifier(n_estimators=600, random_state=14)

# 训练数据
rf.fit(X_train, y_train)
```

```python
# 预测测试集
y_pred = rf.predict(X_test)

# 计算精度
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("./star_train.csv")
df = df.drop("Unnamed: 0",axis=1)
df = df.drop("Unnamed: 0.1",axis=1)

# 定义特征和目标
X = df.drop("star_level", axis=1)
X = X.drop("uid", axis=1)
y = df["star_level"]
print(y.unique())
print(y)


# 分割数据集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=14)
```

```python
params = {
    'booster': 'gbtree',
    'objective': 'multi:softmax',
    'num_class': 10,
    'max_depth': 10,
    'eta': '0.5',
    'eval_metric': 'merror',
    'learning_rate': 0.1, 
    'n_estimators': 500, 
    'max_depth': 5, 
    'min_child_weight': 1, 
    'seed': 0,         
    'subsample': 0.8, 
    'colsample_bytree': 0.8, 
    'gamma': 0,
    'reg_alpha': 0, 
    'reg_lambda': 1
}
```

```python
dtrain = xgb.DMatrix(X_train,label = y_train)
dtest = xgb.DMatrix(X_test,label = y_test)
```

```python
num_rounds = 100
model = xgb.train(params,dtrain,num_rounds,evals=[(dtest,'test')])
```

```python
# 预测测试集
y_pred = model.predict(dtest)
print(y_pred)
# 计算精度
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
model.save_model('./model/xgboost_model_star.bin')
```
