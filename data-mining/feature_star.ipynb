```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./star_final.csv")
```

```python
corr_matrix = df.corr()
corr_matrix
```

```python
corr_matrix.to_csv("./corr_matrix_star.csv")
```

```python
corr_series = corr_matrix.mask(np.triu(np.ones(corr_matrix.shape)).astype(np.bool)).stack()
corr_series
```

```python
to_drop = corr_series[abs(corr_series) >= 0.7].index.tolist()
```

```python
to_drop
drop = []
index = 0
for t in to_drop:
    if (t[0][0:4] == 'MEAN' and t[1][0:4] == 'MAX(') or (t[1][0:4] == 'MEAN' and t[0][0:4] == 'MAX('):
        if t[1][0:4] == 'MAX(':
            to_drop.pop(index)
            drop.append(t[1])
        else:
            to_drop.pop(index)
            drop.append(t[0])
    if (t[0][0:4] == 'MEAN' and t[1][0:4] == 'MIN(') or (t[1][0:4] == 'MEAN' and t[0][0:4] == 'MIN('):
        if t[1][0:4] == 'MIN(':
            drop.append(t[1])
            to_drop.pop(index)
        else:
            drop.append(t[0])
            to_drop.pop(index)
    index = index + 1
```

```python
drop
```

```python
print(len(drop))
```

```python
df.drop(columns=drop,axis=1,inplace=True)
df.to_csv("./star_after_corr.csv")
```

```python
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv("./star_after_corr.csv")
```

```python
df
```

```python
df.drop(columns='Unnamed: 0',axis=1,inplace=True)
x_list = df.columns.to_list()
print(x_list)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

x_list.remove('Unnamed: 0.1')
x_list.remove('uid')
x_list.remove('star_level')
X = df[x_list]
Y = df['star_level']
model = sm.OLS(Y, X).fit()
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
```

```python
print(vif)
```

```python
vif.to_csv("./star_vif_1.csv")
```

```python
vif_threshold = 10
high_vif_features = list(vif[vif['VIF Factor'] > vif_threshold]["features"])
high_vif_features
```

```python
print(len(high_vif_features))
```

```python
df.drop(columns=high_vif_features,axis=1,inplace=True)
```

```python
df.to_csv('./star_after_vif.csv')
```
